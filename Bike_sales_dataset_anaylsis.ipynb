```python
import pandas as pd
```

```python
Bike_sales = pd.read_csv('Bike Sales.csv')
```

```python
Bike_sales
```

```python
Bike_sales.info()
```

```python
# Q-1: What is the total revenue generated per year?
```

```python
# Total revenue per year

total_revenue_per_year = Bike_sales.groupby('Year')['Revenue'].sum()
print(total_revenue_per_year)
```

```python
# Q-2: Which product category has the highest profit margin?
```

```python
# Average profit margin by product category

Bike_sales['Profit_Margin'] = Bike_sales['Profit'] / Bike_sales['Revenue']
avg_profit_margin = Bike_sales.groupby('Product_Category')['Profit_Margin'].mean()

print(avg_profit_margin)
```

```python
# Q-3: What is the age group distribution of customers for each product category?
```

```python
# Age group distribution per product category

age_group_distribution = Bike_sales.groupby(['Product_Category', 'Age_Group']).size()
print(age_group_distribution)
```

```python
# Q-4: Which country has the highest average order quantity?
```

```python
# Average order quantity by country

avg_order_quantity_by_country = Bike_sales.groupby('Country')['Order_Quantity'].mean()
print(avg_order_quantity_by_country)
```

```python
# Q-5: What is the monthly revenue trend?
```

```python
# Monthly revenue trend

monthly_revenue = Bike_sales.groupby('Month')['Revenue'].sum()
print(monthly_revenue)
```
